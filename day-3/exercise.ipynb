```python
import math

mass = int(input('Masukkan Massa (kg) : '))
height = int(input('Masukkan Tinggi (cm) : '))
print(f'Massa {mass} kg dan tinggi {height / 100} m')

IMT = mass / math.pow((height / 100), 2)

if IMT < 18.5 :
    print(f'IMT = {IMT}, BERAT BADAN KURANG')
elif IMT >= 18.5 and IMT <= 24.9 :
    print(f'IMT = {IMT}, BERAT BADAN IDEAL')
elif IMT >= 25.0 and IMT <= 29.9 :
    print(f'IMT = {IMT}, BERAT BADAN BERLEBIH')
elif IMT >= 30.0 and IMT <= 39.9 :
    print(f'IMT = {IMT}, BERAT BADAN SANGAT BERLEBIH')
else :
    print(f'IMT = {IMT}, OBESITAS')
```

```python
stars = ''
column = 5
row = 9
for i in range(row) :
    for j in range(column - i) :
        stars += '* '
    stars += '\n'
    if i == row - column :
        break
for k in range(row - column) :
    for l in range(2 + k):
        stars += '* '
    stars += '\n'
print(stars)
```

```python
# Mendefinisikan harga buah
price_apel = 10000
price_jeruk = 15000
price_anggur = 20000

# Mendefinisikan stock buah
stock_apel = 7
stock_jeruk = 7
stock_anggur = 6

# Meminta input jumlah apel yang dibeli kepada user
while True:
    n_apel = int(input('Masukkan jumlah Apel: '))
    if n_apel > stock_apel :
        print(f"""Jumlah yang dimasukkan terlalu banyak
Stock Apel tinggal: {stock_apel}"""'\n')
        continue
    else:
        break

# Meminta input jumlah jeruk yang dibeli kepada user
while True:
    n_jeruk = int(input('Masukkan jumlah Jeruk: '))
    if n_apel > stock_apel :
        print(f"""Jumlah yang dimasukkan terlalu banyak
Stock Jeruk tinggal: {stock_jeruk}"""'\n')
        continue
    else:
        break
# Meminta input jumlah anggur yang dibeli kepada user
    
while True:
    n_anggur = int(input('Masukkan jumlah Anggur: '))
    if n_apel > stock_apel :
        print(f"""Jumlah yang dimasukkan terlalu banyak
Stock Anggur tinggal: {stock_anggur}"""'\n')
        continue
    else:
        break

# Menghitung total belanjaan per buah
total_apel = n_apel * price_apel
total_jeruk = n_jeruk * price_jeruk
total_anggur = n_anggur * price_anggur

# Menghitung total belanjaan
total = total_apel + total_jeruk + total_anggur

# Memperlihatkan detail belanjaan
print(f"""Detail Belanjaan
      
Apel : {n_apel} x {price_apel} = {total_apel}
Jeruk : {n_jeruk} x {price_jeruk} = {total_jeruk}
Anggur : {n_anggur} x {price_anggur} = {total_anggur}
      
Total : {total}"""'\n')

# Melakukan pembayaran
while True :
    n_uang = int(input('Masukkan jumlah uang : '))
    if n_uang < total :
        print(f"""Uang anda kurang sebesar {total - n_uang}"""'\n')
        continue
    else:
        print(f'''Terima kasih
              
Uang kembali anda : {total - n_uang}''')
        break
```
